```python
# Read Image with pillow
from PIL import Image
img_0 = Image.open("../data/unzipped/burrito_1000/burrito _%d.jpg" % 0)

img_0
```

```python
img_0.size
```

```python
img_0 = img_0.resize((200, 80))
img_0
```

### Cropping Image

```python
# Cropping parts of image
x_up_left = np.random.randint(0, img_0.size[0] // 2)
y_up_left = np.random.randint(0, img_0.size[1] // 2)
x_low_right = np.random.randint(img_0.size[0] // 2, img_0.size[0])
y_low_right = np.random.randint(img_0.size[1] // 2, img_0.size[1])

img_0.crop((x_up_left,
            y_up_left,
            x_low_right,
            y_low_right)).resize((200, 80))
```

### Rotating Image

```python
img_0.rotate(np.random.randint(360))
```

### Transposing Image

```python
img_0.transpose(Image.FLIP_LEFT_RIGHT)
```

### Drawing Shapes on Image

```python
from PIL import ImageDraw

# Coordinates of ellipse
x_up_left = np.random.randint(0, img_0.size[0] // 2)
y_up_left = np.random.randint(0, img_0.size[1] // 2)
x_low_right = np.random.randint(x_up_left, img_0.size[0])
y_low_right = np.random.randint(y_up_left, img_0.size[1])

img_0_copy = img_0.copy()

draw_img_0 = ImageDraw.Draw(img_0_copy)

draw_img_0.ellipse([x_up_left,
                    y_up_left,
                    x_low_right,
                    y_low_right], fill='brown')
img_0_copy
```

```python
# Coordinates of rectangle
x_up_left = np.random.randint(0, img_0.size[0] // 2)
y_up_left = np.random.randint(0, img_0.size[1] // 2)
x_low_right = np.random.randint(x_up_left, img_0.size[0])
y_low_right = np.random.randint(y_up_left, img_0.size[1])

img_0_copy = img_0.copy()

draw_img_0 = ImageDraw.Draw(img_0_copy)

draw_img_0.rectangle([x_up_left,
                      y_up_left,
                      x_low_right,
                      y_low_right], fill='blue')
img_0_copy
```

```python
# Lines
x_up_left = np.random.randint(0, img_0.size[0] // 2)
y_up_left = np.random.randint(0, img_0.size[1] // 2)

img_0_copy = img_0.copy()

draw_img_0 = ImageDraw.Draw(img_0_copy)

for i in range(np.random.randint(1000)):
    draw_img_0.line([(x_up_left, y_up_left),(i - x_up_left, i - y_up_left)], fill='green')
        
img_0_copy
```
